```python
import numpy as np
from matplotlib import pyplot as plt
from math import tan, sin, cos, radians
```

```python
alpha = 30
m = 0.05
v = 10
g = 9.81
D = 0.507
```

```python
def dragless(t):
    x = v*t*cos(radians(alpha))
    y = v*t*sin(radians(alpha)) - (g*t*t)/2
    return x,y

dlx = [0]
dly = [0]

for t in np.linspace(0, 150, 2000):
    dragless_x, dragless_y = dragless(t)

    if dragless_y < 0:
        continue

    dlx.append(dragless_x)
    dly.append(dragless_y) 

plt.plot(dlx,dly)
plt.show()
```
